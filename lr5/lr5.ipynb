```python
sentence = "Положив трубку, Тэнго сел за стол, включил словопроцессор и начал переносить в электронную память все, что написал чернилами на бумаге в гостинице Тикуры."
```

```python
# токенизация

from spacy.lang.ru import Russian
import spacy

nlp = spacy.load('ru_core_news_sm')
spacy_text = nlp(sentence)
spacy_text
```

```python
# частеречная разметка
for token in spacy_text:
    print('{} - {} - {}'.format(token.text, token.pos_, token.dep_))
```

```python
# лемматизация
for token in spacy_text:
      print(token, token.lemma, token.lemma_)
```

```python
# выделение (распознавание) именованных сущностей
for ent in spacy_text.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
displacy.render(spacy_text, style='ent', jupyter=True)
```

```python
from spacy import displacy
displacy.render(spacy_text, style='dep', jupyter=True)
```
